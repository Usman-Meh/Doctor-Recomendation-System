# **Cleaning DataSet**

**Cleaning Experience column**

```python
type(df['Experience'][0])
```

```python
# Replace the pattern with an empty string
df['Experience'] = df['Experience'].where(df['Experience'].str.contains('Years', case=False), None)
```

```python
pattern = r'Years'

# Replace the pattern with an empty string
df['Experience'] = df['Experience'].str.replace(pattern, '', regex=True)
```

```python
try:
  # Attempt conversion with coercion (replacing non-numeric values with NaN)
  df['Experience'] = pd.to_numeric(df['Experience'], errors='coerce')
except pd.errors.ParserError:
  # Handle potential parsing errors (e.g., invalid formats)
  print("Error: Some values could not be converted to integers.")
  # You might choose to keep the original data or perform alternative actions here
```

```python
df['Experience'].isnull().sum()
```

```python
E_value=int(df['Experience'].mean())
```

```python
df['Experience'].dtype
```

```python
df['Experience'].fillna(value=E_value,inplace=True)
```

**Cleaning Fee column**

```python
df.columns
```

```python
type(df['Fee'][0])
```

```python
# Replace the pattern with an empty string
df['Fee'] = df['Fee'].where(df['Fee'].str.contains('Rs', case=False), None)
df['Fee']
```

```python
pattern_2 = r'Rs.'

# Replace the pattern with an empty string
df['Fee'] = df['Fee'].str.replace(pattern_2, '', regex=True)
df['Fee']
```

```python
df['Fee'] = pd.to_numeric(df['Fee'].str.replace(',', ''), errors='coerce')
```

```python
df['Fee'].dtype
```

```python
df['Fee'].isnull().sum()
df['Fee']
```

```python
df['Fee'].fillna(value=1000,inplace=True)
```

**Cleaning Number of patients**

```python
df
```

```python
df['Number of Patients:']
```

```python
specific_string = 'not available'

# Replace the specific string with None
df['Number of Patients:'] = df['Number of Patients:'].replace(specific_string, None)
```

```python
df['Number of Patients:'] = pd.to_numeric(df['Number of Patients:'].str.replace(',', ''), errors='coerce')
```

```python
N_value = int(df['Number of Patients:'].std())
```

```python
df['Number of Patients:'].fillna(value=N_value,inplace=True)
```

```python
df['Number of Patients:']
```

***Cleaning Waiting time column***

```python
# Replace the pattern with an empty string
df['Wait_time'] = df['Wait_time'].where(df['Wait_time'].str.contains('Min', case=False), None)
df['Wait_time'].value_counts()
```

```python
df['Wait_time'].fillna(value='15 - 30 Min',inplace=True)
```

```python
df['Wait_time'].value_counts()
```

**Cleaning Highest Degree and Degrees column**


```python
H_value=df['Highest degree: '].mode()
H_value
```

```python
df['Highest degree: '].fillna(value='MBBS',inplace=True)
# Standardize 'M.B.B.S' and 'M.B.B.S.' to 'MBBS' (replace with your desired column name)
df['Highest degree: '] = df['Highest degree: '].str.replace(r'(M\.B\.B\.S\.?)', 'MBBS', regex=True)
df['Highest degree: '] = df['Highest degree: '].str.replace(r'(B\.D\.S\.?)', 'MBBS', regex=True)

df['Highest degree: '].value_counts()
```

```python
df['Degrees:'].ffill(inplace=True)
```

**Cleaning Rating column**

```python
df.dropna(inplace=True)
```

```python
df['Rating'] = pd.to_numeric(df['Rating'].str.replace(',', ''), errors='coerce')
```

# **Data Preprocessing**

```python
import pandas as pd
df1 = pd.read_csv('/content/drive/MyDrive/Recomendation system/Cleaned oladoc file.csv')
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df2=df1.copy()
df2.drop(['Name: '],inplace=True,axis=1)
```

**Creating "satisfaction level" column**

```python
#As rating is already in 100% syntax so we take it as it is
df2.rename(columns={'Rating':'Satisfaction_level'},inplace=True)
```

**Creating Labels**

```python
#As in document we have to create labels in which satisfaction level is above 95% will be 1 and otherwise 0
df2['recommended'] = df2['Satisfaction_level'] > 95
df2['recommended'] = df2['recommended'].astype(int)
df2['recommended'].value_counts()
```

```python
df2.isnull().sum()
```

```python
df2.describe()
```

```python
df2.info()
```

```python
df2.drop(['Degrees:'],axis=1,inplace=True)
```

```python
X =df2.iloc[:,:-1]
Y =df2.iloc[:,-1]
Y
```

```python
df2.columns
```

**Encoding and Scalling**

```python
import sklearn.preprocessing as pre_process
import numpy as np
ordinal_encoding =pre_process.OrdinalEncoder()
standered_scalling  =pre_process.StandardScaler()
one_hot =pre_process.OneHotEncoder()

from sklearn.compose import make_column_transformer
transform_x= make_column_transformer( (one_hot , ['Specialization', 'Highest degree: ','Wait_time','Location'])
,(standered_scalling,['Experience','Fee', 'Number of Patients:', 'Satisfaction_level']))
```

```python
transform_x
```

```python
processed_x= transform_x.fit_transform(X)
processed_y= ordinal_encoding.fit_transform(np.c_[Y])
```

# **Spliting DataSet**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(processed_x, processed_y, test_size=0.25, random_state=42)
```

**Import Model**

```python
from sklearn.ensemble import RandomForestClassifier
# Initialize Random Forest classifier
RF_algorithm = RandomForestClassifier(n_estimators=100, random_state=42)
```

**Train the Model**

```python
RF_algorithm.fit(X_train,y_train)
```

```python
# Predict y label through model
y_pred = RF_algorithm.predict(X_test)
```

**Accuracy and Confusion_Matrix**

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
