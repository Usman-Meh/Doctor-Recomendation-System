```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python

    
count=10
while count<=100:
        a=str(count)
        b='/'
        c=b+a
    url= "https://oladoc.com/pakistan/lahore/gynecologist"
    response = requests.get(url)
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')


    # Find all elements representing each doctor box
    doctor_boxes = soup.find_all(class_='card-body p-lg-4 doctor-listing-card')
    #print(doctor_boxes)


    Name = []
    Degree = []
    Details = []
    Experience = []
    Fee = []
    Wait_time =[]
    Number_of_patients =[]
    Rating =[]
    # Extract and print information for each doctor
    for doctor_box in doctor_boxes:
        # Extract doctor name
        name = doctor_box.find(class_='doctor-name').text.strip()
        
        # Extract doctor degree
        degree = doctor_box.find(class_='doc-degree').text.strip()
        
        # Extract doctor details
        details = doctor_box.find(class_='doc-specialization').text.strip()
        
        
        # Extract doctor's fee
        doctor_fee_tag = doctor_box.find('span', class_='doctor-fee')
        doctor_fee = doctor_fee_tag.text.strip() if doctor_fee_tag else 'Not Available'
        
        # Extract number of ratings
        ratings_elem = doctor_box.find(class_='d-inline-block font-weight-normal')
        ratings = ratings_elem.text.strip() if ratings_elem else "not available"
        
        
        wait_time = "not available"
        # Extract doctor experience and waiting time
        items = doctor_box.find_all(class_='item')
        if len(items) >= 2:
            wait_time_elem = items[0].find(class_='od-wte')
            experience_elem = items[1].find(class_='od-wte')
            if wait_time_elem:
                wait_time = wait_time_elem.text.strip()
                
            if experience_elem:
                experience = experience_elem.text.strip()
            if len(items) > 2:
                rating_percentage = [item.find(class_='od-wte').text.strip().split('\n')[0] for item in items[2:] if item.find(class_='od-wte')]
        
        # Print extracted information
    #     print("Name:", name)
    #     print("Degree:", degree)
    #     print("Specialization:", details)
    #     print("Fee:", doctor_fee)
    #     print('Number of Patients: ',ratings)
    #     print("Waiting Time:", wait_time)
    #     print("Experience:", experience)
    #     print('Rating: ',wait_times)
    #     print("---"*20)
        
        # Append extracted information to lists
        Name.append(name)
        Degree.append(degree)
        Details.append(details)
        Experience.append(experience)
        Fee.append(doctor_fee)
        Number_of_patients.append(ratings)
        Wait_time.append(wait_time)
        Rating.append(rating_percentage)

        count+=10
        
        
```

```python
rating = [inner[0] if inner else None for inner in Rating]
# Splitting degrees into two columns
highest_degree = [deg.split(',')[0].strip() for deg in Degree]
degree_name = [', '.join(deg.split(',')[1:]).strip() if ',' in deg else '' for deg in Degree]
```

```python
dic={'Name: ':Name,
     'Specialization': Details,
     'Highest degree: ':highest_degree,
     'Experience':Experience,
     'Fee':Fee,'Degrees:':degree_name,
     'Wait_time':Wait_time,
     'Number of Patients:':Number_of_patients,
     'Rating':rating,
     'Location':'Lahore'
    }
df =pd.DataFrame(dic)
df['Number of Patients:'] = df['Number of Patients:'].str.strip('()')

df
```

```python
df.to_csv('oldoc_data.csv',mode='a',index=False,header=False)
```
